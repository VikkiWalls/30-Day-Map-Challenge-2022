```python
wind = gpd.read_file("ukcp18-uk-land-12km-wind.shp")
```

```python
print(wind.crs)
```

```python
land = gpd.read_file("ukcp18-uk-land-country-hires.shp")
```

```python
print(land.crs)
```

Building the map with Geopandas

```python
ax1 = wind.plot(figsize=(50,50), cmap='plasma', alpha=1, legend=True)
land.plot(figsize=(50,50), color = 'cyan', ax = ax1, alpha = 0.3)
ax1.set_axis_off()
```

Experimenting with an alternate colour scheme and layer orientation

```python
ax =  land.plot(figsize=(50,50), color = 'navy')
wind.plot(figsize=(50,50), cmap='plasma', alpha=0.45, ax=ax, legend=True)
ax.set_axis_off()
```

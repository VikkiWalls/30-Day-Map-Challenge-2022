```python
import earthpy.spatial as es
import matplotlib.pyplot as plt
import os
import rasterio

from matplotlib import colors
```

```python
os.getcwd()
os.chdir("c:\\Users\\vicks\\OneDrive\\Data Science (not uni)\\Portfolio\\30 Day Map Challenge\\30 Day Map Challenge Data\\Island")
```

Importing the [Data](https://earthexplorer.usgs.gov/scene/metadata/full/5e83a1f36d8572da/GMTED2010N10W030/) (Clipped and Gaussian Filtered using QGIS)

```python
file = rasterio.open("la gomera clip gaus.tif")
```

```python
lagomera = file.read()
```

```python
print(lagomera.shape)
```

Initial inspection of the raster file

```python
plt.imshow(lagomera[0], cmap='Spectral')
plt.show()
```

Creating a hillshade plot to overlay onto the map - producing the 3D effect

```python
hillshade = es.hillshade(lagomera[0], azimuth=270, altitude=1)
```

Creating the map

```python
fig, ax = plt.subplots(figsize=(20, 20))
i = plt.imshow(lagomera[0], cmap='plasma_r', norm=colors.LogNorm())
ax.imshow(hillshade, cmap="Greys", alpha=0.2)
ax.axis('off')

txt = ax.text(0.02, 0.03, "La Gomera",
              size=36,
              color='#630FB9',
              transform = ax.transAxes,
              fontfamily='fantasy')

plt.show()
```

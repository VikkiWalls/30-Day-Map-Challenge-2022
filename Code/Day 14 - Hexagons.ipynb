Importing the [data](https://www.gbif.org/occurrence/search?offset=80&dataset_key=0a0d4ca1-5bfd-4594-9d51-40b972308bf5&has_coordinate=true&has_geospatial_issue=false&geometry=POLYGON((-21.64723%2050.24421,14.64723%2050.24421,14.64723%2064.75579,-21.64723%2064.75579,-21.64723%2050.24421))&occurrence_status=present)

```python
bee = pd.read_csv("Bee locations.csv")
bee.head()
```

Building the hexbin map

```python
fig = ff.create_hexbin_mapbox(
    data_frame=bee, lat="Latitude", lon="Longitude",
    nx_hexagon=20, opacity=0.5, labels={"color": "Bee Sightings"},
    zoom = 7.5,
    color_continuous_scale="YlOrBr",
    min_count=1,
    show_original_data=True,
    original_data_marker=dict(size=4, opacity=0.6, color="gold")
)
fig.update_layout(margin={"r": 0, "t": 0, "l": 0, "b": 0}, mapbox_style="mapbox://styles/mapbox/dark-v10", paper_bgcolor = 'rgba(0,0,0,0.85)')
fig.update_coloraxes(colorbar_bgcolor='rgba(0,0,0,0.85)', colorbar_xpad=10, colorbar_tickfont_color='gold',
                        colorbar_title_font_color='gold', cmin=0)
fig.show()
```

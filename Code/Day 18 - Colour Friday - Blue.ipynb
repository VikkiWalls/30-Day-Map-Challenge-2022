Importing Tuna [Data](https://mola.stanford.edu/NaturePaper/vr-index.php)

```python
tuna = pd.read_csv("Tuna 510010700 migration.csv")
```

```python
tuna.head()
```

```python
countries = gpd.read_file("ne_10m_admin_0_countries.shp")
```

```python
print(countries.crs)
```

```python
tuna = gpd.GeoDataFrame(tuna, geometry=gpd.points_from_xy(tuna.mplon, tuna.mplat))
```

```python
tuna.crs = {'init': 'epsg:4326'}
```

```python
tuna.head()
```

```python
ax = countries.plot(figsize=(15,15), color='darkgreen')
tuna.plot( marker='o', color='coral', markersize=2, ax=ax)
ax.set_facecolor("darkblue")
ax.set_axis_off();
```

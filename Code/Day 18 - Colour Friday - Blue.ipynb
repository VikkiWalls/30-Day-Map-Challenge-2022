```python
import os
import pandas as pd
import plotly.graph_objects as go
import plotly.express as px

%matplotlib inline
px.defaults.width = 1090
px.defaults.height = 600
```

Setting the working Directory

```python
os.getcwd()
os.chdir("c:\\Users\\vicks\\OneDrive\\Data Science (not uni)\\Portfolio\\30 Day Map Challenge\\30 Day Map Challenge Data\\Blue")
```

Importing Tuna [Data](https://mola.stanford.edu/NaturePaper/vr-index.php)

```python
tuna1 = pd.read_csv("Tuna 510010700 migration.csv")
tuna1.head()
```

```python
tuna2 = pd.read_csv("Tuna 510010800 migration.csv")
```

```python
tuna3 = pd.read_csv("Tuna 510011700 migration.csv")
```

```python
tuna4 = pd.read_csv("Tuna 510011900 migration.csv")
```

```python
tuna5 = pd.read_csv("Tuna 510012700 migration.csv")
```

Importing my mapbox token

```python
px.set_mapbox_access_token(open('mapboxtokenblue.txt').read())
```

Building map

```python
fig = px.line_mapbox(tuna1, lat="mplat", lon="mplon")
fig.add_trace(go.Scattermapbox(
    mode = "lines",
    lon = tuna1.iloc[:,9],
    lat = tuna1.iloc[:,10],
    marker = dict(size= 10, color = '#5855ff'),
    name = "Tuna 1"))

fig.add_trace(go.Scattermapbox(
    mode = "lines",
    lon = tuna2.iloc[:,9],
    lat = tuna2.iloc[:,10],
    marker = dict(size= 10, color = '#2636f6'),
    name = "Tuna 2"))

fig.add_trace(go.Scattermapbox(
    mode = "lines",
    lon = tuna3.iloc[:,9],
    lat = tuna3.iloc[:,10],
    marker = dict(size= 10, color = '#1976d2'),
    name = "Tuna 3"))

fig.add_trace(go.Scattermapbox(
    mode = "lines",
    lon = tuna4.iloc[:,9],
    lat = tuna4.iloc[:,10],
    marker = dict(size= 10, color = '#03a9f4'),
    name = "Tuna 4"))

fig.add_trace(go.Scattermapbox(
    mode = "lines",
    lon = tuna5.iloc[:,9],
    lat = tuna5.iloc[:,10],
    marker = dict(size= 10, color = '#4dd0e1'),
    name = "Tuna 5"))

fig.update_layout(mapbox_style="mapbox://styles/datasciencenot2/clamk221f002q14qum1znx73t", mapbox_zoom=3.5, mapbox_center_lat = 34,mapbox_center_lon = -50,
    margin={"r":0,"t":0,"l":0,"b":0},paper_bgcolor = 'rgba(1,19,86,100)')

fig.show()
```

```python
import os
import pandas as pd
import plotly.figure_factory as ff
import plotly.express as px

%matplotlib inline
px.defaults.width = 1070
px.defaults.height = 600
```

Importing the [data](https://github.com/VikkiWalls/30-Day-Map-Challenge-2022/blob/main/Self%20Gathered%20Data/Dolly%20Parton%20Stage%20Visits.csv) - concert details found [here](https://www.concertarchives.org/bands/dolly-parton--2)

```python
dolly = pd.read_csv("Dolly Parton Stage Visits to 2022.csv")
dolly.head()
```

Building a hexabin map

```python
fig = ff.create_hexbin_mapbox(
    data_frame=dolly, lat="Lat", lon="Long",
    nx_hexagon=40, opacity=0.5, labels={"color": "Dolly Concerts"},
    zoom = 2,
    color_continuous_scale=['#5a007d', '#8d0ebf', '#c122fe', '#e900ff', '#f696fe', '#f9cdfd'],
    min_count=1,
    show_original_data=True,
    original_data_marker=dict(size=4, opacity=0.6, color="#f9cdfd")
)
fig.update_layout(margin={"r": 0, "t": 0, "l": 0, "b": 0}, mapbox_style="mapbox://styles/mapbox/dark-v10", paper_bgcolor = 'rgba(0,0,0,0.85)')
fig.update_coloraxes(colorbar_bgcolor='rgba(0,0,0,0.85)', colorbar_xpad=10, colorbar_tickfont_color='#e900ff',
                        colorbar_title_font_color='#e900ff', cmin=0)
fig.show()
```

```python
fig.write_html("dollyplot2022.html")
```

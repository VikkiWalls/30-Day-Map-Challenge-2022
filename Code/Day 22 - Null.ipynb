## Day 22 - Null

Plan: Use OpenStreetMap to map Nulltown, Indiana

```python
import osmnx as ox
```

Selecting roads of Nulltown (Indiana) using coordinates (obtained by viewing OpenStreetMap Export page)

```python
Streets = ox.graph_from_bbox(39.6098,39.5499,-85.2216,-85.0950,retain_all=True, truncate_by_edge=True,
                        simplify=False, network_type='all')
```

Plotting the map

```python
fig, ax = ox.plot_graph(Streets,figsize=(50,50), edge_color='#FFFFFF', bgcolor='black', node_size=0.1, edge_linewidth=3)
```

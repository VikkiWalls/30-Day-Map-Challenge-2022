# 30 Day Map Challenge

## Day 5 - Ukraine

Plan: Map the Chornobyl site and the cool meanders in the Prypiat River

```python
import osmnx as ox
```

Importing the data - from [OpenStreetMap](https://www.openstreetmap.org/export#map=12/51.2428/30.3399)

```python
Streets = ox.graph_from_bbox(51.3251,51.2279,30.1446,30.3978,retain_all=True, truncate_by_edge=True,
                        simplify=False, network_type='drive')
```

```python
Water = ox.graph_from_bbox(51.3251,51.2279,30.1446,30.3978, retain_all=True, truncate_by_edge=True,
                        simplify=False, custom_filter='["natural"~"water"]')
```

Plotting the edges of all the waterways

```python
fig, ax = ox.plot_graph(Water,figsize=(50,50), edge_color='gold', bgcolor='navy', node_size=0.1)
```

Plotting the streets

```python
ox.plot_graph(Streets,figsize=(50,50), edge_color='gold', bgcolor='black',  node_size=5, node_alpha=0.5)
```

```python
import matplotlib.pyplot as plt
import os
import subprocess

from osgeo import gdal
```

Setting the working directory

```python
os.getcwd()
os.chdir("c:\\Users\\vicks\\OneDrive\\Data Science (not uni)\\Portfolio\\30 Day Map Challenge\\30 Day Map Challenge Data\\Raster")
```

Building a shell command to compute the DEM slope into a new file

```python
slpcmd = "gdaldem slope s30_e029_1arc_v3.tif slope.tif -compute_edges"
```

Running the command

```python
os.system(slpcmd)
```

Building an array of the slope data

```python
slp = gdal.Open("slope.tif")
```

```python
slpArray = slp.GetRasterBand(1).ReadAsArray()
```

Visualising the array

```python
plt.figure(figsize = (20,20))
plt.imshow(slpArray)
plt.colorbar()
plt.title("Slope Plot ", fontsize = 40, pad='40.0')
plt.show()
```

Building a shell command to compute the DEM aspect into a new file

```python
aspcmd = "gdaldem aspect s30_e029_1arc_v3.tif aspect.tif -compute_edges"
```

Running the command

```python
os.system(aspcmd)
```

Building an array of the aspect data

```python
asp = gdal.Open("aspect.tif")
```

```python
aspArray = asp.GetRasterBand(1).ReadAsArray()
```

Visualising the array

```python
plt.figure(figsize = (20,20))
plt.imshow(aspArray, cmap = "magma")
plt.colorbar()
plt.title("Aspect Plot ", fontsize = 40, pad='40.0')
plt.show()
```

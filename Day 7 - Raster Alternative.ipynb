```python
import matplotlib.pyplot as plt
import os
import subprocess

from osgeo import gdal
gdal.UseExceptions()
```

Setting the working directory

```python
os.getcwd()
os.chdir("c:\\Users\\vicks\\OneDrive\\Data Science (not uni)\\Portfolio\\30 Day Map Challenge\\30 Day Map Challenge Data\\Raster")
```

Building a shell command to compute the DEM slope

```python
cmd = "gdaldem slope s30_e029_1arc_v3.tif slope1.tif -compute_edges"
```

Creating a new slope raster file

```python
os.system(cmd)
```

Building an Array of the slope data

```python
slp1 = gdal.Open("slope1.tif")
```

```python
slp1Array = slp1.GetRasterBand(1).ReadAsArray()
```

Visualising the array

```python
plt.figure()
plt.imshow(slp1Array)
plt.colorbar()
plt.show()
```
